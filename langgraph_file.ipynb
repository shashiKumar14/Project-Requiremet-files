```python
State :1. Shared datastructure that holds the current information or
        context of the entire application
        2. In simple terms ,it is like the applications memory ,keeping track of the variables
        and data that nodes can access and modify as they execute
```

```python
Nodes : Nodes are individual functions or operations that
 perform specific tasks within the graph
```

```python
Graph : A Graph in Langgraph is the overorching structure that
maps out how differnt  tasks (nodes) are connected and executed
```

```python
Edges: Connection between the nodes and determine the 
flow of execution
```

```python
Conditional edge: are specialized connections that decide the next node to execute
based on specific conditions or logic applied to the current state
```

```python
ToolNode: is just aspecial kind of node whose main job 
is to run a tools.
It connects the tools output back into the state,
so other nodes can use that information
```

```python
StateGraph:is a class in langgraph ,used to build and compile the graph structure.
```

```python
Runnable: in langgraph is a standardized,executable coponent
that perform a specific task within an ai workflow.

it serves as a fundamental building block 
allowing for us to create modular systems
```
