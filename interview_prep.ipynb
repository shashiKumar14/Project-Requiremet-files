```python
s = "{[()]}"
#s="{[(])}"
stack = []
mapping = {')':'(', '}':'{', ']':'['}

for char in s:
    if char in mapping.values():
        stack.append(char)
    elif char in mapping:
        if not stack or stack.pop() != mapping:
            print("Not Balanced")
            break
else:
    print("Balanced" if not stack else "Not Balanced")
```

```python
s = "{[()]}"
stack = []
pairs = {')':'(', ']':'[', '}':'{'}
i = 0

while i < len(s):
    if s[i] in pairs.values():
        stack.append(s[i])
    elif not stack or stack.pop() != pairs[s[i]]:
        break
    i += 1

print("Balanced" if i == len(s) and not stack else "Not Balanced")
```
